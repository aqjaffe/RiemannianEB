```python
import os

def print_directory_structure(path='.', prefix='', max_depth=3, current_depth=0):
    """Print the directory structure of a folder"""
    if current_depth >= max_depth:
        return
    
    try:
        items = sorted(os.listdir(path))
    except PermissionError:
        print(f"{prefix}[Permission Denied]")
        return
    
    for i, item in enumerate(items):
        item_path = os.path.join(path, item)
        is_last = i == len(items) - 1
        
        current_prefix = '└── ' if is_last else '├── '
        print(f"{prefix}{current_prefix}{item}")
        
        if os.path.isdir(item_path):
            extension_prefix = '    ' if is_last else '│   '
            print_directory_structure(item_path, prefix + extension_prefix, max_depth, current_depth + 1)

# Print structure of current directory
print_directory_structure()
```
