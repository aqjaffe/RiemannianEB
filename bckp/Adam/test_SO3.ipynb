# $\textrm{SO}(3)$ Simulation

```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
from geomstats.geometry.special_orthogonal import _SpecialOrthogonal3Vectors
import geomstats.visualization as visualization
```

```python
plt.rcParams.update({'font.size': 10,
                     'mathtext.fontset': 'stix',
                     'font.family': 'serif',
                     'font.serif':'Palatino'})
```

## Setting up

```python
SO3 = _SpecialOrthogonal3Vectors()
SO3
```

```python
n = 1000
mean = SO3.random_uniform()
# mean = np.eye(3)
sigma2 = 0.1
```

```python
X = np.asarray([SO3.projection(np.random.normal(mean,np.sqrt(sigma2),size=(3,3))) for i in range(n)])
X_angs = SO3.tait_bryan_angles_from_matrix(X, extrinsic=True, zyx=True)
X_angs.shape
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12,3), sharex=True,sharey=True)

axs[0].hist(X_angs[:,0])
axs[0].set_xlim([-np.pi, np.pi])
axs[0].set_xlabel('$\\alpha$')
axs[1].hist(X_angs[:,1])
axs[1].set_xlabel('$\\beta$')
axs[2].hist(X_angs[:,2])
axs[2].set_xlabel('$\\gamma$')

plt.suptitle('Riemannian Gaussian in $SO(3)$ (Tait-Bryan Euler angles)')
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X_angs[:,0], X_angs[:,1], X_angs[:,2], alpha=0.2)

ax.set_xlim([-np.pi, np.pi])
ax.set_xlabel('$\\alpha$')
ax.set_ylim([-np.pi/2, np.pi/2])
ax.set_ylabel('$\\beta$')
ax.set_zlim([-np.pi, np.pi])
ax.set_zlabel('$\\gamma$')

plt.title('Riemannian Gaussian in $SO(3)$\n(Tait-Bryan Euler angles)')
```
